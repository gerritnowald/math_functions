```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reproducible random points
np.random.seed(0)
x = np.linspace(0, 9, 10)  # 10 x values
a = 2.5  # slope
b = 1.0  # intercept
noise = np.random.normal(scale=2.0, size=x.shape)  # gaussian noise
y = a * x + b + noise
```

```python
Φ1 = (sum(x)*sum(y) - len(x)*sum(x*y))/((sum(x))**2 - len(x)*sum(x**2))
Φ0 = (sum(y) - Φ1*sum(x))/len(x)

print((Φ0, Φ1))
```

```python
xplot = np.linspace(0,9)
yfit  = Φ0 + Φ1 * xplot

# Plot points, true line, and fitted line
plt.scatter(x, y, label='data')
plt.plot(xplot, yfit, 'r', label='fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
