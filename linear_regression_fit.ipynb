```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reproducible random points
x = np.linspace(0, 9, 10)  # 10 x values
a = 0.5  # slope
b = 3.0  # intercept

rand = np.random.Generator(np.random.PCG64(0))
noise = rand.normal(scale=0.5, size=x.shape)  # gaussian noise

y = a * x + b + noise
```

The linear regression function  
$y_i = \phi_0 + \phi_1 x_i$

has the cost function (least squares)  
$L(\phi) = \sum_{i=1}^{N} (\phi_0 + \phi_1 x_i - y_i)^2$ .

At the minimum of $L$, its partial derivatives are 0:  
$\frac{\partial L}{\partial\phi_0} = \sum (\phi_0 + \phi_1 x_i - y_i)$ = 0  
$\frac{\partial L}{\partial\phi_1} = \sum (\phi_0 + \phi_1 x_i - y_i) x_i$ = 0

which gives  
$\phi_0 N + \phi_1\sum x_i - \sum y_i = 0$  
$\phi_0\sum x_i + \phi_1\sum x_i^2 - \sum x_i y_i = 0$ .

These two equations can be solved for $\phi_0, \phi_1$:

```python
Φ1 = (sum(x)*sum(y) - len(x)*sum(x*y))/((sum(x))**2 - len(x)*sum(x**2))
Φ0 = (sum(y) - Φ1*sum(x))/len(x)

print((Φ1, Φ0))
```

```python
xplot = np.linspace(0,9)
yfit  = Φ0 + Φ1 * xplot

# Plot points, true line, and fitted line
plt.scatter(x, y, label='data')
plt.plot(xplot, yfit, 'r', label='fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
